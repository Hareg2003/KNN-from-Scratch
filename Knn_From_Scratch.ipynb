## Separating Categorical and Numerical Columns:

```python
X_train.head()
```

```python
X_train_cat = X_train.select_dtypes(include=['object'])

X_train_cat.head()
```

```python
X_train_num = X_train.select_dtypes(include=['int64', 'float64'])

X_train_num.head()
```

## Applying Label Encoding on Categorical Columns

```python
X_train_cat_le = pd.DataFrame(index=X_train_cat.index)

X_train_cat_le.head()
```

```python
X_train_cat.cut.unique()
```

```python
cut_encoder = {'Fair' : 1, 'Good' : 2, 'Very Good' : 3, 'Ideal' : 4, 'Premium' : 5}

X_train_cat_le['cut'] = X_train_cat['cut'].apply(lambda x : cut_encoder[x])

X_train_cat_le.head()
```

```python
X_train_cat.color.unique()
```

```python
color_encoder = {'J':1, 'I':2, 'H':3, 'G':4, 'F':5, 'E':6, 'D':7}

X_train_cat_le['color'] = X_train_cat['color'].apply(lambda x : color_encoder[x])

X_train_cat_le.head()
```

```python
X_train_cat.clarity.unique()
```

```python
clarity_encoder = {'I1':1, 'SI2':2, 'SI1':3, 'VS2':4, 'VS1':5, 'VVS2':6, 'VVS1':7, 'IF':8}

X_train_cat_le['clarity'] = X_train_cat['clarity'].apply(lambda x : clarity_encoder[x])

X_train_cat_le.head()
```

## Scaling the Numerical Features

```python
X_train_num.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_num_rescaled = pd.DataFrame(scaler.fit_transform(X_train_num), 
                                    columns = X_train_num.columns, 
                                    index = X_train_num.index)

X_train_num_rescaled.head()
```

## Concatinating the Encoded Categorical Features and Scaled Numerical Features

```python
X_train_transformed = pd.concat([X_train_num_rescaled, X_train_cat_le], axis=1)

X_train_transformed.head()
```

## Method 1
### Using Enumerate function, indexing location wise

```python
Y_predict_1=[]
dist_list=[]
X_test_idx=np.array(X_train_transformed.index)
for index,i in tqdm(enumerate(X_test_transformed.index)):
    dist=(((X_train_transformed-X_test_transformed.loc[i])**2).sum(axis=1))**0.5
    dist_list.append((dist))
    sort_dist=np.argsort(dist_list[index])
    sort_y=X_test_idx[sort_dist]
    y_index=sort_y[:5]
    y_nn=y_train[y_index]
    Y_predict_1.append(np.mean(y_nn))
```

```python
y_df=pd.DataFrame({"y_test":y_test,"Y_predict":Y_predict_1})
y_df.head()
```

```python
from sklearn import metrics

print('Mean Absolute Error: ', metrics.mean_absolute_error(y_test, Y_predict_1))
```

## Method 2
### Taking range of length of X_test_transformed

```python
Y_predict=[]
dist_list=[]
X_test_idx=np.array(X_train_transformed.index)
for i in tqdm(range(len(X_test_transformed))):
    dist=np.sqrt((np.square(X_train_transformed.values-X_test_transformed.values[i])).sum(axis=1))
    dist_list.append(dist)
    sort_dist=np.argsort(dist_list[i])
    sort_y=X_test_idx[sort_dist]
    y_index=sort_y[:5]
    y_nn=y_train[y_index]
    Y_predict.append(np.mean(y_nn))
```

```python
y_df=pd.DataFrame({"y_test":y_test,"Y_predict":Y_predict})
y_df
```

```python
from sklearn import metrics

print('Mean Absolute Error: ', metrics.mean_absolute_error(y_test, Y_predict))
```

# KNN with Sklearn

```python
from sklearn.neighbors import KNeighborsRegressor
regressor = KNeighborsRegressor()
regressor.fit(X_train_transformed, y_train)
```

```python
y_test_pred = regressor.predict(X_test_transformed)
```

```python
temp_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_test_pred})

temp_df.head()
```

```python
from sklearn import metrics

print('Mean Absolute Error: ', metrics.mean_absolute_error(y_test, y_test_pred))

print('Mean Squared Error: ', metrics.mean_squared_error(y_test, y_test_pred))

print('Root Mean Squared Error: ', np.sqrt(metrics.mean_squared_error(y_test, y_test_pred)))
```
